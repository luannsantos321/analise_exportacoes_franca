### Puxar base de dados

```python
import pandas as pd

exportacoes = pd.read_csv('exportacoes_franca.csv')
```

### Informações Gerais

```python
display(exportacoes)
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

```python
france = exportacoes.loc[exportacoes['Country']== 'France']
ano = france[['Year', 'SH4 Code']]
ano_2016 = ano.loc[ano['Year'] == 2016].value_counts()
ano_2017 = ano.loc[ano['Year'] == 2017].value_counts()
ano_2018 = ano.loc[ano['Year'] == 2018].value_counts()
ano_2019 = ano.loc[ano['Year'] == 2019].value_counts()
ano_2020 = ano.loc[ano['Year'] == 2020].value_counts()
print(ano_2016)
print(ano_2017)
print(ano_2018)
print(ano_2019)
print(ano_2020)
```

### Quais os produtos mais exportados ao longo de todo o período?
